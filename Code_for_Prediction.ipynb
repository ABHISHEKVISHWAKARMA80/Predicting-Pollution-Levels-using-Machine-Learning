```python
def hypothesis(X,theta):
    return np.dot(X,theta)
def error(X,Y,theta):
    error=0.0
    m=X.shape[0]
    y_=hypothesis(X,theta)
    e=np.sum((Y-y_)**2)
    return e/m
def gradient(X,Y,theta):
    y_=hypothesis(X,theta)
    grad=np.dot(X.T,(y_ - Y))
    m=X.shape[0]
    return grad/m
def gradient_descent(X,Y,learning_rate=0.1,max_epochs=150):
    n=X.shape[1]
    print(n)
    theta=np.zeros((n,))
    print(theta)
    error_list=[]
    print(error_list)
    for i in range(max_epochs):
        
        e=error(X,Y,theta)
        error_list.append(e)
        
        grad=gradient(X,Y,theta)
        theta=theta-learning_rate*grad
    return theta,error_list
```

```python
theta,error_list=gradient_descent(X,Y)
print(theta)
```

```python
print(error_list)
```

```python
y_test=hypothesis(x_test,theta)
print(y_test)
```

```python
df=pd.DataFrame(data=y_test,columns=["target"])
print(df)
```
